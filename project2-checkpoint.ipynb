```python
def malli(plot,title,xlabel,ylabel):
    plot.set_xlabel(xlabel,fontsize=16,weight="bold")
    plot.set_ylabel(ylabel,fontsize=13,weight="bold")
    plot.set_title(title,fontsize=15,weight="bold")
```

```python
plt.figure(figsize=(15,6))
sns.set_style("whitegrid")
ax=sns.countplot(data=tv,x="brand",color="orange")
malli(ax,"vizualize the brands sales","products","sales")
ax.tick_params("x",rotation=90)
for i in ax.containers:
    ax.bar_label(i)
```
